Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

```python
class Solution:
    def containsDuplicates(self,arr:list[int])->bool:
        hashset = set()
        for n in arr:
            if n in hashset:
                return True
            hashset.add(n)
        return False
```

```python
result = Solution()
```
