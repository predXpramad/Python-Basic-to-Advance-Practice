#### SQL And SQLite
SQL (Structured Query Language) is a standard language for managing and manipulating relational databases. SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems. In this lesson, we will cover the basics of SQL and SQLite, including creating databases, tables, and performing various SQL operations.

````python
import sqlite3
````

````python
## Connect to SQLLite Database

connection = sqlite3.connect('example.db')
connection
````

````python
# Cusrsor Object is used to iteravte over rows and colums
cursor = connection.cursor()
````

````python
## Create Table
cursor.execute('''
CREATE TABLE IF NOT EXISTS employees(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    department text
)
'''
)

## Commit the changes
connection.commit()
````

````python
cursor.execute('''
SELECT * FROM employees
'''
)
````

````python
## Insert Data in SQLite table
cursor.execute('''
INSERT INTO EMPLOYEES(name, age,department)
               VALUES('Krish',32,'AD')
'''
)

cursor.execute('''
INSERT INTO EMPLOYEES(name, age,department)
               VALUES('Manoj',32,'ISE')
'''
)

cursor.execute('''
INSERT INTO EMPLOYEES(name, age,department)
               VALUES('Raj',33,'CSE')
'''
)

cursor.execute('''
INSERT INTO EMPLOYEES(name, age,department)
               VALUES('Vinay',30,'EEE')
'''
)

cursor.execute('''
INSERT INTO EMPLOYEES(name, age,department)
               VALUES('Rahul',33,'ME')
'''
)

## Commit changes to DB
connection.commit()
````

````python
# Query Data through SQLite DB

cursor.execute('''
SELECT * FROM employees
'''
)

rows = cursor.fetchall()

# print the quried data

for row in rows:
    print(row)

````

````python
## Update data in table

cursor.execute('''
UPDATE employees
SET age = 34
WHERE age == 33 or name == 'Manoj'
'''
)

connection.commit()
````

````python
# Query Data through SQLite DB

cursor.execute('''
SELECT * FROM employees
'''
)

rows = cursor.fetchall()

# print the quried data

for row in rows:
    print(row)

````

````python
# Delete the data from the table

cursor.execute('''
DELETE FROM employees
WHERE name == 'Raj'
'''
)

connection.commit()
````

````python
## Working with Sales Data, Like a Mini-project

## connect to an SQLite Database
salesDatabase = sqlite3.connect('Sales_data.db')
cursor = salesDatabase.cursor()

# Create a table for sales data
cursor.execute(
'''
CREATE TABLE IF NOT EXISTS sales(
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT

) 
'''
)


sales_data = [
    ('2023-01-01', 'Product1', 100, 'North'),
    ('2023-01-02', 'Product2', 200, 'South'),
    ('2023-01-03', 'Product1', 150, 'East'),
    ('2023-01-04', 'Product3', 250, 'West'),
    ('2023-01-05', 'Product2', 300, 'North')
]

## BULK INSERT using 'executemany'
cursor.executemany('''
INSERT INTO sales(date, product,sales,region)
VALUES(?,?,?,?)

''',sales_data
)


salesDatabase.commit()
````

````python
cursor.execute(
'''
SELECT *
FROM sales

'''
)

rows = cursor.fetchall()

for row in rows:
    print(row)
````

````python
## closes the connect with database

salesDatabase.close()
````
