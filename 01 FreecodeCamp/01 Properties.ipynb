Properties are a type of Decorator function that allows us to run methods like attributes without using `()` or `semi-collon`. 

They lets us run logic behind the scenes while getting or setting attributes values.

Properties ties getters and setters together so you can execute logic while maintaining dot notation access?

The below two decorators are used to create getters and setters for a property -
* @property
* @<property_name>.setter

Finally they are used to retrive, update and modify values of an attribute.

```python
## Always remeber to use Single Underscore when using property, otherwise it goes into recursion
class Circle:
    def __init__(self, radius):
        self.radius = radius

    @property
    def radius(self):
        return self._radius 
    
    @radius.setter # When ever function is called it automatically validates values in condition
    def radius(self,value):
        if value < 0:
            raise ValueError("Value Cannot be zero")
        self._radius = value

    @radius.deleter
    def radius(self):
        print("Deleting Radius..")
        del self._radius
        print("Radius Deleted!")
```

```python
shape = Circle(4)
print(shape.radius)
```

```python
shape.radius = 10
print(shape.radius)
```

```python
del shape.radius
```
