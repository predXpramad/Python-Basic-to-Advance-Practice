A setter is a method used to set the value of an attribute, allowing for validation checks and restrictions. You can create a setter using the `@propertyName.setter` decorator, where `propertyName` should match the name of the property to set:


As you learned in a previous lesson, a setter offers a way to control how an attribute can be modified. To ensure that `new_name` is the right type, create an `if` statement that raises a `TypeError` with the message `'name' must be a string`. when n`ew_name` is not an instance of `str`.

```python
class Employee:
    _base_salaries = {
        'trainee': 1000,
        'junior': 2000,
        'mid-level': 3000,
        'senior': 4000
    }
    def __init__(self,name,level):
        if not (isinstance(name,str) and isinstance(level,str)):
            raise TypeError("'name' and 'level' attribute must be of type 'str'.")
        if level not in self._base_salaries:
            raise ValueError(f"Invalid value '{level}' for 'level' attribute.")

        
        self._name = name
        self._level = level
        self._salary = self._base_salaries[self.level]
    
    def __str__(self):
        return f"{self.name}: {self.level}"
    
    @property
    def name(self):
        return self._name
    
    @name.setter
    def name(self, new_name):
        if not isinstance(new_name,str):
            raise TypeError("'name' must be a string.")
        self._name = new_name
        print(f"'name' updated to '{self._name}'.")
    
    @property
    def level(self):
        return self._level
    
    @level.setter
    def level(self, new_level):
        if new_level not in Employee._base_salaries:
            raise ValueError(f"Invalid value '{new_level}' for 'level' attribute.")
        # Added check for Step 27
        if new_level == self.level:
            raise ValueError(f"'{self.level}' is already the selected level.")
        # if Employee._base_salaries.get(new_level) < Employee._base_salaries.get(self._level):
        #     raise ValueError("Cannot change to lower level.")
        # OR
        if Employee._base_salaries[new_level] < Employee._base_salaries[self.level]:
            raise ValueError(f"Cannot change to lower level.")
        # Access salary setter, rather than changing salary directly
        self.salary = Employee._base_salaries[new_level]    
        self._level = new_level
        print(f"'{self._name}' promoted to '{new_level}'.")
    
    @property
    def salary(self):
        return self._salary

    @salary.setter
    def salary(self,new_salary):
        if not (isinstance(new_salary,int) or isinstance(new_salary,float)):
            raise TypeError("'salary' must be a number.")
        if new_salary < self._salary:
            raise ValueError(f"Salary must be higher than minimum salary ${self._salary}.")
        self._salary = new_salary
        print(f"Salary updated to ${self._salary}")

    def __repr__(self):
        return f"Employee('{self.name}', '{self.level}')"
    
charlie_brown = Employee("Charlie Brown","trainee")
print(charlie_brown)
print(repr(charlie_brown))
print(f"Base salary: ${charlie_brown.salary}")
charlie_brown.name = "Charlie Brown Junior"
charlie_brown.name = "Charlie Brown"

charlie_brown.level = "trainee"
print(charlie_brown.level)
```
