```python
import seaborn as sns
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
```

```python
print(type(tips))
tips.head()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter Plot Total bill v/s Tip")
plt.show()
```

```python
tips.head()
```

```python
# Line Plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.show()
```

```python
## Categorical Plots (--> uses categorical Variables)

# Bar Plot
sns.barplot(data=tips,x='day',y='total_bill')
plt.title('Bar Plot of total bill by day')
plt.show()
```

```python
## Box Plot 
plt.figure(figsize=(11,11))
plt.subplot(2,2,1)
sns.boxplot(data = tips)
plt.title("'tips' Dataset")


plt.subplot(2,2,2)
sns.boxplot(data = tips,x='total_bill')
plt.title("Box Plot of Total bills")

plt.subplot(2,2,3)
sns.boxplot(data = tips, x = 'tip')
plt.title("'tips' Dataset")

plt.subplot(2,2,4)
sns.boxplot(data = tips, x = 'size')
plt.title("")

plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
sns.boxplot(x='day', y = 'total_bill', data=tips)
plt.subplot(1,2,2)
sns.boxplot(x='time', y = 'total_bill', data=tips)
```

```python
sns.violinplot(x='day', y = 'total_bill', data=tips)
```

```python
## Histograms

sns.histplot(tips['total_bill'],bins=10, kde=True, color='black')
```

```python
## KDE Plot Kernel Density Estimation

sns.kdeplot(tips['total_bill'])
```

```python
## KDE Plot Kernel Density Estimation with shade

sns.kdeplot(tips['total_bill'], fill = True)
plt.show()
```

```python
sns.pairplot(data=tips)
```

```python
tips.head()
```

```python
## HeatMaps
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr , annot=True, cmap='Oranges')
```

```python
import pandas as pd

df = pd.read_csv('sales_data.csv')
df.head()
# type(df)
```

```python
df.pivot_table(index='Product Category', columns='Region',values='Total Revenue',aggfunc='sum')
```

```python
df.pivot_table(index='Product Category', columns='Payment Method',values='Total Revenue',aggfunc='sum')
```

```python
df.head()
```

```python
# Plot total Sales by Product

plt.figure(figsize=(10,5))
sns.barplot(x = 'Product Category', y = 'Total Revenue', data = df) # Without estimator, default value is mean in that bin
plt.title('Sales by product Category')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Plot total Sales by Product

plt.figure(figsize=(10,5))
sns.barplot(x = 'Product Category', y = 'Total Revenue', data = df, estimator=sum)
plt.title('Total Sales by product Category')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Plot total Sales by Region

plt.figure(figsize=(10,5))
sns.barplot(x = 'Region', y = 'Total Revenue', data = df, estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
